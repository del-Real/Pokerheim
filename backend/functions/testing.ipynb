```python
import requests
```

```python
# create request to create a table
base_url = "http://127.0.0.1:5001/pokergame-007/us-central1/"

def send_request(endpoint, params=None):
    url = f"{base_url}{endpoint}"
    if params:
        url += "?" + "&".join([f"{key}={value}" for key, value in params.items()])
    response = requests.post(url)
    print(f"{response.status_code} {response.text}")
```

```python
send_request("createTable", {"tableId": "table_1"})
send_request("joinTable", {"tableId": "table_1", "playerId": "player_1"})
send_request("joinTable", {"tableId": "table_1", "playerId": "player_2"})
send_request("joinTable", {"tableId": "table_1", "playerId": "player_3"})

send_request("playerStatus", {"playerId": "player_1", "status": "ready"})
send_request("playerStatus", {"playerId": "player_2", "status": "ready"})

send_request("startGame", {"tableId": "table_1"})
```

```python
send_request("performAction", {"playerId": "player_1", "action": "bet", "amount": 20})
send_request("performAction", {"playerId": "player_2", "action": "bet", "amount": 50})
send_request("performAction", {"playerId": "player_1", "action": "call"})
send_request("performAction", {"playerId": "player_2", "action": "bet", "amount": 50})
send_request("performAction", {"playerId": "player_1", "action": "call"})
send_request("performAction", {"playerId": "player_2", "action": "fold"})
```

```python
for i in range(50):
    send_request("createTable", {"tableId": f"table_{i}"})
    for j in range(3):  
        send_request("joinTable", {"tableId": f"table_{i}", "playerId": f"player_{i}_{j}"})
        send_request("playerStatus", {"playerId": f"player_{i}_{j}", "status": "ready"})
        send_request("startGame", {"tableId": f"table_{i}"})
```
